```python
import pandas as pd
import numpy as np
import collections
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from nltk.tokenize import word_tokenize
```

```python
S1= 'sunshine state enjoy sunshine'
S2 ='brown fox jump high, brown fox run'
S3 ='sunshine state fox run fast'
```

```python
from sklearn.feature_extraction.text import CountVectorizer     
vectorizer = CountVectorizer()
                                                                # Bag Of Words
Model = vectorizer.fit_transform([S1,S2,S3])
bagofwords = pd.DataFrame(Model.toarray(),columns=vectorizer.get_feature_names_out())
bagofwords
```

```python
dictionary_S1 = bagofwords.iloc[0].to_dict()
bowS1 = S1.split(" ")

dictionary_S2 = bagofwords.iloc[1].to_dict()        
bowS2 = S2.split(" ")

dictionary_S3 = bagofwords.iloc[2].to_dict()
bowS3 = S3.split(" ")

print("S1 =", dictionary_S1 , "\n", bowS1 , "\n")
print("S2 =", dictionary_S2 , "\n", bowS2 , "\n")
print("S3 =", dictionary_S3 , "\n", bowS3 , "\n")
```

```python
def TF(wordDict, bow):
    tfDict = {}
    bowCount = len(bow)
    for word, count in wordDict.items():
        tfDict[word] = count/float(bowCount)      # Tf Func. & Table
    return tfDict

tf_of_S1 = TF(dictionary_S1, bowS1)
tf_of_S2 = TF(dictionary_S2, bowS2)
tf_of_S3 = TF(dictionary_S3, bowS3)

tf=pd.DataFrame.from_dict([tf_of_S1 , tf_of_S2 , tf_of_S3])
tf
```

```python
def IDF(docList):
    import math
    idfDict = {}
    N = len(docList)
    
    idfDict = dict.fromkeys(docList[0].keys(), 0)     # Idf 
    for doc in docList:
        for word, val in doc.items():
            if val > 0:
                idfDict[word] += 1
    
    for word, val in idfDict.items():
        idfDict[word] = math.log10(N / float(val))
        
    return idfDict  

idf = IDF([dictionary_S1,dictionary_S2,dictionary_S3])
idf_pd=pd.DataFrame([idf], index=['idf'])
idf_pd  
```

```python
def TFIDF(tfBow, idfs):
    tfidf = {}
    for word, val in tfBow.items():
        tfidf[word] = val*idfs[word]
    return tfidf                              #Tf_IDF Func. & Table

tfidf_S1 = TFIDF(tf_of_S1, idf)
tfidf_S2 = TFIDF(tf_of_S2, idf)
tfidf_S3 = TFIDF(tf_of_S3, idf)

tfid_pd= pd.DataFrame([tfidf_S1, tfidf_S2,tfidf_S3], index=['S1','S2','S3'])
tfid_pd
```

```python
# Q.2 Cosine Similarity

import numpy as np
from numpy.linalg import norm
 
S1 = np.array([0,	1,	0,	0,	0,	0,	0,	1,	2])                 
S3 = np.array([0,	0,	1,	1,	0,	0,	1,	1,	1])

print("S1:", S1)
print("S3:", S3)
 
cosine = np.dot(S1,S3)/(norm(S1)*norm(S3))
print("\n Cosine Similarity:", cosine)
```
